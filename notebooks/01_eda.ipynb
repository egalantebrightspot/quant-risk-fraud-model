# 01 – Exploratory Data Analysis

EDA for risk and fraud modeling: load data, **class imbalance**, and a quick feature glimpse. Runs top-to-bottom (use synthetic data if CSV missing).

```python
# Project root for imports (run from notebooks/ or project root)
import sys
from pathlib import Path
ROOT = Path().resolve().parent if Path().resolve().name == "notebooks" else Path().resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.config import DATA_DIR, RANDOM_STATE
from src.data.schema import TARGET_FRAUD
from src.data.loaders import generate_and_save

sns.set_theme(style="whitegrid")
%matplotlib inline
DATA_PATH = ROOT / "data" / "training_data.csv"
```

```python
# Load data: use CSV if present, else generate synthetic (reproducible)
if not DATA_PATH.exists():
    print("No CSV found; generating synthetic data...")
    df = generate_and_save(path=DATA_PATH, n=25_000, fraud_rate=0.05, random_state=RANDOM_STATE)
else:
    df = pd.read_csv(DATA_PATH)
print(f"Shape: {df.shape[0]:,} rows, {df.shape[1]} columns")
df.head()
```

## Class imbalance view

Target distribution (fraud_flag: 0 = non-fraud, 1 = fraud). Critical for calibration and evaluation.

```python
# Class counts and rate
counts = df[TARGET_FRAUD].value_counts().sort_index()
rate = df[TARGET_FRAUD].mean()
print(counts.to_string())
print(f"\nFraud rate: {rate:.2%}")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
counts.plot(kind="bar", ax=axes[0], color=["#2ecc71", "#e74c3c"], edgecolor="black")
axes[0].set_title("Class counts")
axes[0].set_xlabel(TARGET_FRAUD)
axes[0].set_xticklabels(["Non-fraud (0)", "Fraud (1)"], rotation=0)
axes[0].set_ylabel("Count")

counts.plot(kind="pie", ax=axes[1], labels=["Non-fraud", "Fraud"], autopct="%1.1f%%", colors=["#2ecc71", "#e74c3c"], startangle=90)
axes[1].set_title("Class balance")
axes[1].set_ylabel("")
plt.tight_layout()
plt.show()
```
